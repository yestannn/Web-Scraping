```
from bs4 import BeautifulSoup
import requests
from pprint import pprint
import pandas as pd
```

```
page = requests.get(
    "https://contentservice.kz/produkty-pitaniya-napitki-kz/")
soup = BeautifulSoup(page.content, 'html.parser')
uls = soup.findAll('ul',{'class':"link-list link-list_md"})
type(uls)
a = uls[0].findAll('a')
hrefs = {}
for i in a:
    hrefs[i.getText()] = (i.get('href'))
```

```
pprint(hrefs)
```

```
roots = {}
for href in hrefs:
    page = requests.get(
    "https://contentservice.kz" + hrefs[href])
    soup = BeautifulSoup(page.content, 'html.parser')
    uls = soup.findAll('ul',{'class':"link-list link-list_md"})
    a = uls[0].findAll('a')
    sub_hrefs = []
    for i in a:
        if href not in roots:
            roots[href] = []
        else:
            roots[href].append(i.get('href'))
    pprint(roots)


```

```
link ='/ryazhenka-6/'
def pages_link(link):
    page = requests.get('https://contentservice.kz' + link)

    soup = BeautifulSoup(page.content, 'html.parser')
    uls = soup.findAll('ul',{'class':"pagination category-pagination"})
    try:
        pages = uls[0].findAll('li')
        pages = pages[1:-1]
        pages_link = []
        for page in pages:
            pages_link.append(page.findAll('a')[0].get('href'))
    except IndexError:
        pages_link = [link]
    return (pages_link)

pprint(pages_link(link))
```

```
count = 0
item_hrefs = []
for root in roots:
    for link in roots[root]:
        print(link)
        pages = pages_link(link)
        for link in pages:
            page = requests.get(
            "https://contentservice.kz" + link)
        
            soup = BeautifulSoup(page.content, 'html.parser')
            uls = soup.findAll('div',{'class':"catalog__grid-view"})
            items = uls[0].findAll('div', {'class':'catalog__grid-view__item'})
            for item in items:
                count += 1
                item_hrefs.append(item.findAll('a')[0].get('href'))

pprint(item_hrefs)
```

```
print(len(item_hrefs))
```

```
data = {
    'name': [],
    'certified': [],
    'company': [],
    'country': [],
    'category_1': [],
    'category_2': [],
    'category_3': [],
    'link': []
}
link =  '/product/2133200000001-kz-maslo-krestyanskoe-rast-sliv-kg'
def product_info(link, data):
    page = requests.get(
            "https://contentservice.kz" + link)
    soup = BeautifulSoup(page.content, 'html.parser')
    try:
        certify = soup.findAll('div', {'class': 'single-product__verification'})[0].getText()
        certified = True
    except:
        certified = False
    if 'Товар не прошел верификацию' in certify:
        certified = False
    else:
        certified = True
    try:
        name = soup.findAll('p', {'class': 'product-specifications-title'})[0].getText()
        name = name[15:]
    except:
        name = None
    try:
        category_1 = soup.findAll('ol', {'class': 'breadcrumb'})[0].findAll('a')[1].getText()
        category_2 = soup.findAll('ol', {'class': 'breadcrumb'})[0].findAll('a')[2].getText()
        category_3 = soup.findAll('ol', {'class': 'breadcrumb'})[0].findAll('a')[3].getText()
    except:
        category_1 = None
        category_2 = None
        category_3 = None
    link_ = "https://contentservice.kz" + link
    try:
        company = soup.findAll('ul', {'class': 'list-unstyled list-lg-pad'})[0].findAll('li')[0].getText() 
        if company == 'Страна ':
            company = None
    except:
        company = None
    try:
        country = soup.findAll('ul', {'class': 'list-unstyled list-lg-pad'})[0].findAll('span')[0].get('title')
    except:
        country = None

    data['name'].append(name)
    data['certified'].append(certified)
    data['category_1'].append(category_1)
    data['category_2'].append(category_2)
    data['category_3'].append(category_3)
    data['company'].append(company)
    data['country'].append(country)
    data['link'].append(link_)


```

```
for link in item_hrefs:
    print(link)
    product_info(link, data)

```

```
df = pd.DataFrame.from_dict(data)
df.to_excel("output.xlsx") 
```
